## Step 2: Preprocess the Data

```python
# Gives no. of classes info, this is needed in 'Dense' layer in the model
import numpy as np
np.unique(ytrain)
```

### 2(i) Scaling the Train and Test data pixel values (0-255)

```python
xtrain_scl = xtrain.astype("float32")/255
xtest_scl = xtest.astype("float32")/255
```

### 2(ii) One Hot Encoding as this is multiclass classification

```python
from keras.utils import to_categorical
ytrain_cat = to_categorical(ytrain)
ytest_cat = to_categorical(ytest)
```

## Step 3: Create CNN Model (Convolutional Neural Network)

image.png

```python
xtrain[0].shape
```

```python
from keras.models import Sequential
from keras.layers import Input, Conv2D, MaxPool2D, Flatten, Dropout, Dense
```

```python
# Initialize the model
model = Sequential()

# Add Input Layer
model.add(Input(shape=(28,28,1)))

# Add Concolution Layer 1
model.add(Conv2D(filters=32, kernel_size=(3,3), activation='relu'))
# Add Max Pool layer 1
model.add(MaxPool2D(pool_size=(2,2)))
# Add Concolution Layer 2
model.add(Conv2D(filters=64, kernel_size=(3,3), activation='relu'))
# Add Max Pool layer 2
model.add(MaxPool2D(pool_size=(2,2)))

# Add Flatten Layer
model.add(Flatten())

# Add Dropout Layer
model.add(Dropout(0.5))

# Add Dense Layer
model.add(Dense(10, activation='softmax'))
```

```python
model.summary()
```

image.png

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Training the model
hist = model.fit(xtrain_scl, ytrain_cat, validation_split=0.1, epochs=20)
```

## Step 4: Plot the Learning Curve

```python
plt.figure(figsize=(10,5))
plt.plot(hist.history['loss'], label='Train Loas')
plt.plot(hist.history['val_loss'], label='Validation Loas')
plt.xlabel('Number of Epochs')
plt.ylabel('Categorical Crossentropy')
plt.title('Learning Curve')
plt.legend()
plt.grid()
plt.show()
```

## Step 5: Evaluate the model

```python
model.evaluate(xtrain_scl, ytrain_cat)
```

```python
model.evaluate(xtest_scl, ytest_cat)
```

## Step 6: Predict results for Train and Test

```python
yprob_train = model.predict(xtrain_scl)
yprob_test = model.predict(xtest_scl)
```

```python
yprob_train[0:5]
```

```python
# get the predictions from probabiltities
ypred_train = np.argmax(yprob_train, axis=1)
ypred_test = np.argmax(yprob_test, axis=1)
```

## Step 7: Plot Confusion Matrix for Test data

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
cf = confusion_matrix(ytest, ypred_test)
cfd = ConfusionMatrixDisplay(cf)
cfd.plot()
```

### 7(i) Print Classification Report

```python
from sklearn.metrics import classification_report
print(classification_report(ytest, ypred_test))
```
