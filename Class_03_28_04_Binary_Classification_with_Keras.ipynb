# Binary Classification - Booking Cancellation

## Step 1: read the dataset

```python
import pandas as pd
url1 = "https://raw.githubusercontent.com/utkarshg1/Neural-Net-Datasets/main/booking_train.csv"
df = pd.read_csv(url1)
df.head()
```

### Booking Status --> Target feature

1. 0 --> Not Cancelled
2. 1 --> Cancelled

## Step 2: Pefrom Basic Data Quality Checks

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

## Step 3: Separate X and Y (booking_status)

```python
X = df.drop(columns=['id', 'booking_status'])
Y = df[['booking_status']]
```

```python
X.head()
```

```python
Y.head()
```

```python
Y.value_counts()
```

```python
Y.value_counts().plot(kind='bar')
```

## Step 4: Apply preprcoessing on X

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
num_pipe = Pipeline(steps=[('impute', SimpleImputer(strategy='median')),
                           ('scaler', StandardScaler())]).set_output(transform='pandas')
```

```python
X_pre = num_pipe.fit_transform(X)
X_pre.head()
```

## Step 5: Perform Train-test Split

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(X_pre, Y, test_size=0.2, random_state=42)
```

```python
xtrain.head()
```

```python
ytrain.head()
```

```python
xtest.head()
```

```python
ytest.head()
```

```python
xtrain.shape
```

```python
xtest.shape
```

## Step 6: Build the Neural Netwrok

image.png

image.png

```python
xtrain.shape[1]
```

```python
from keras.models import Sequential
from keras.layers import Input, Dense, Dropout
```

```python
# Initialize the model
model = Sequential()

# Add Input layer
model.add(Input(shape=(xtrain.shape[1])))

# Add Hidden layer 1
model.add(Dense(units=64, activation='relu'))
# Add Dropout Layer 1
model.add(Dropout(rate=0.4))

# Add Hidden layer 2
model.add(Dense(units=32, activation='relu'))
# Add Dropout Layer 2
model.add(Dropout(rate=0.3))

# Add Output layer
model.add(Dense(units=1, activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
hist = model.fit(xtrain, ytrain, validation_split=0.2, epochs=60)
```

## Step 7: Plot the Learning Curve

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 5))
plt.plot(hist.history['loss'], label='Train Loss')
plt.plot(hist.history['val_loss'], label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Binary Crossentropy')
plt.title('Learning Curve')
plt.legend()
plt.grid()
plt.show()
```

## Step 8: Evalaute model in Train and Test

```python
model.evaluate(xtrain, ytrain)
```

```python
model.evaluate(xtest, ytest)
```

## Step 9: Predict the Results for Train & Test

```python
# Activation Function 'sigmoid' gives probabilty predictions. SO the below will give these values
yprob_train = model.predict(xtrain)
yprob_test = model.predict(xtest)
```

```python
yprob_train[0:5]
```

```python
yprob_test[0:5]
```

```python
# Converting the predicted probabilties to Prediction (0 & 1)
ypred_train = [1 if prob>=0.5 else 0 for prob in yprob_train]
ypred_test = [1 if prob>=0.5 else 0 for prob in yprob_test]
```

```python
ypred_train[0:5]
```

```python
ytrain.head()
```

```python
ypred_test[0:5]
```

## Step 10: Plot Confusion Matrix --> This step performed for Classification Target feature

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
cf = confusion_matrix(ytest, ypred_test)
cfd = ConfusionMatrixDisplay(cf)
cfd.plot()
```

#### From above matric Tue(0) and Predicted(1) --> 4412 + 2371 --> this is good model

```python
# Classification report
from sklearn.metrics import classification_report
print(classification_report(ytest, ypred_test))
```

### CONCLUSION: Based on above report --> the model is good as f1-accuracy = 0.81 which is >0.8
